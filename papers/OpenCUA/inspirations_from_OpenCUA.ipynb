# Inspiration for DifyCAD

## Inspiration1. Reflector是核心灵魂

数据质量＞数据数量，要用Reflector保证数据质量。
1. 不要盲目收集CATIA操作数据
2. 必须用LLM分析每一步：这步对吗？为什么这么做？有更好的方法吗？
3. 生成带反思的CoT推理: 
    - task
    - history: step, action
    - current_state:
        - screen_shots: screen_shots1, screen_shots2, screen_shots3
        - reasoning: memory, reflection, plan, prediction
        - action: description, code

标注流程：
1. 用户自然操作CATIA
2. 完成后写任务描述
3. 工具自动处理成轨迹
4. 人工审核质量：ok/good/excellent

错误恢复机制：错误不可怕，关键是能识别和纠正
1. 每3-5步自我检查：我完成了吗？方向对吗？有没有偏离目标
2. 如果发现错误：回溯到正确的步骤 or 用新方案替代

## Inspiration2. 两阶段训练是标准范式

1. Stage-1：
    - 收集CATIA界面标注：工具栏、菜单、按钮的位置
    - 混合通用CAD知识
    - COM接口状态
2. Stage-2：
    - 收集CAD&KBE工作流轨迹
    - 用LLM生成反思CoT
    - 保留一些Stage-1的数据（20%）

## Inspiration3. 对历史表示的设计很关键

1. 视觉：保留3张CATIA界面截图（两步之前、上一步、当前）
2. 文本：简洁的动作序列
    ```text
    Step1：click（拉伸按钮）
    Step2：dragTo（草图平面）
    Step3：write("50mm")
    ```
3. 当前步骤加Memory：
    ```text
    前面创建了一个圆形草图，
    现在要拉伸50mm形成圆柱体
    ```

## Inspiration4. 混合通用数据很重要

训练数据配比：
1. 50% CATIA操作轨迹：CAD工作流，带Reflector生成的CoT
2. 20% CATIA界面定位：工具栏按钮位置、菜单选项位置、手动标注边界框
3. 20% CAD通用知识：机械制图规范、GD&T标注、工程材料知识、CATIA教程文档
4. 10% 通用能力数据：中文指令理解、数学计算、逻辑推理

## Inspiration5. Pass@N显示巨大潜力

可参考的策略：多方案生成（比单次生成可靠）
1. 生成多个方案（N=3-5）
2. 用规则筛选：
    - 步数最少的
    - 最符合工程规范的
    - 最容易执行的
3. 让用户选择 or 用reranking模型自动选最优
